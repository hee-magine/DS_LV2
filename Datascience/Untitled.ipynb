```python
df_q2 = df_q2.reset_index(drop = True).fillna(0)
df_q2.head(2)
```

```python
df_q2["prod_cat"] = df_q2["prod_cat1"].astype("int").astype("str") + "-" \
+ df_q2["prod_cat2"].astype("int").astype("str") + "-" \
+ df_q2["prod_cat3"].astype("int").astype("str")
df_q2.head(10)
```

```python
df_q2_m0 = df_q2.loc[df_q2["marital"] == 0, ]
df_q2_m1 = df_q2.loc[df_q2["marital"] == 1, ]
```

```python
mean_m0 = df_q2_m0.groupby("user")["prod_cat"].nunique().mean() # number of unique value (elemnet)
mean_m1 = df_q2_m1.groupby("user")["prod_cat"].nunique().mean()
```
