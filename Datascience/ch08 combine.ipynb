```python
# 2번 문제
bike.info()
```

```python
bike['datetime'] = pd.to_datetime(bike['datetime'])
bike['hour'] = bike['datetime'].dt.hour
bike.head(3)
```

```python
bike_s2 = bike.loc[bike['season'] == 2, ] # 여름 추출
bike_s4 = bike.loc[bike['season'] == 4, ] # 겨울 추출
```

```python
bike_s2_hour = bike_s2.groupby('hour')['registered'].mean().reset_index()
bike_s2_hour = bike_s2_hour.rename(columns = {'registered':'registered_s2'})
bike_s2_hour.head(3)
```

```python
bike_s4_hour = bike_s4.groupby('hour')['registered'].mean().reset_index()
bike_s4_hour = bike_s4_hour.rename(columns = {'registered':'registered_s4'})
bike_s4_hour.head(3)
```

```python
bike_join = pd.merge(left = bike_s2_hour, right = bike_s4_hour,
                   left_on = 'hour', right_on = 'hour',
                   how = 'left')
bike_join.head(3)
```

```python
bike_join['diff'] = abs(bike_join['registered_s2'] - bike_join['registered_s4'])
bike_join.head(3)
```

```python
bike_join.loc[bike_join['diff'] == bike_join['diff'].max(), ]
```

```python
# 3번 문제, 다시 데이터 불러오기 부터
bike = pd.read_csv('bike.csv')
bike.head(3)
```

```python
bike['datetime'] = pd.to_datetime(bike['datetime'])
bike.info()
```

```python
bike['date'] = bike['datetime'].dt.date
bike['hour'] = bike['datetime'].dt.hour
bike.head()
```

```python
bike_hum = bike.groupby('date')['humidity'].max().reset_index()
bike_hum = bike_hum.loc[bike_hum['humidity'] == 100]
bike_hum.head()
```

```python
bike_tgt = bike.iloc[:, [12, 13, 5, 11]]
bike_tgt.head()
```

```python
bike_tgt = pd.merge(left = bike_tgt, right = bike_hum,
                    left_on = 'date', right_on = 'date',
                    how = 'inner')
bike_tgt.head()
```

```python
bike_tgt = bike_tgt.loc[bike_tgt['temp'] > 30]
bike_tgt['count'].mean()
```
